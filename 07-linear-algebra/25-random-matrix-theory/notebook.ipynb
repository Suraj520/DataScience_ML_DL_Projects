#### About

> Random Matrix Theory

Random Matrix Theory (RMT) is a branch of mathematics that studies the statistical properties of matrices with random entries. In RMT, the entries of a matrix are assumed to be random variables with certain statistical properties, such as Gaussian distributions. 

RMT has applications in many fields, such as quantum mechanics, wireless communications, and machine learning.

One common use case of RMT is in analyzing the performance of random matrix-based algorithms in machine learning, such as principal component analysis (PCA) and random projection. RMT provides a theoretical framework for understanding the behavior of these algorithms and predicting their performance under different scenarios.





```python
import numpy as np
from scipy.stats import wishart
```

```python
# Generate a random matrix with Wishart distribution
n = 1000 # dimension of matrix
p = 500 # number of observations
X = np.random.normal(size=(n, p))
C = np.dot(X.T, X)
W = wishart.rvs(df=n, scale=C, size=1)
```

```python
# Compute the eigenvalues of the matrix
eigvals = np.linalg.eigvalsh(W)
```

```python
# Plot the distribution of eigenvalues
import matplotlib.pyplot as plt
plt.hist(eigvals, bins=50)
plt.title("Distribution of Eigenvalues")
plt.show()
```
