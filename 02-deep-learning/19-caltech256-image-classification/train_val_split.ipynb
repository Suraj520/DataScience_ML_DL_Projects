This notebook is aimed towards creating a test and train split from identical copies of dataset containing 30k images one of which is renamed as train folder and its copy as val folder inside a root directory Caltech256

Dataset Link - https://www.kaggle.com/datasets/jessicali9530/caltech256

```python
import os
import glob
import numpy as np
```

```python
root_dir= "/home/suraj/ClickUp/Jan-Feb/data/Caltech256"
```

```python
#deleting last 10 images from each folder in train data
for filepath in glob.glob(root_dir+"/train/*"):
    file_indices=[]
    
    # finding list of last 10 images from each folder that needs to be deleted from train set 
    for file in glob.glob(filepath+"/*.jpg"):
        file_idx = file.split('/')[-1].split('.')[0].split('_')[1]
        meta_data = file.split('/')[-1].split('.')[0].split('_')[0]
        file_indices.append(int(file_idx))
    #finding the range of file indices to delete
    idxs_to_del_start = (max(file_indices)-10)
    idxs_to_del_end = (max(file_indices))
    #deleting the files
    indices_to_del = np.arange(idxs_to_del_start,idxs_to_del_end+1)
    #changing precision to have leading zeros as per file naming convention 
    indices_to_del = [format(num, "04") for num in indices_to_del]
    #iterating through each and deleting them
    for img_idx in indices_to_del:
        img_name = filepath+ "/"+str(meta_data)+"_"+str(img_idx)+".jpg"
        #deleting the file
        try:
            os.remove(img_name)
        except FileNotFoundError:
            print("{} not found".format(img_name))
    print("{} processed".format(filepath))
    
```

```python
#keep last 10 images from each folder in val data and deleting rest
for filepath in glob.glob(root_dir+"/val/*"):
    file_indices=[]
    
    # finding list of last 10 images from each folder that needs to be deleted from train set 
    for file in glob.glob(filepath+"/*.jpg"):
        file_idx = file.split('/')[-1].split('.')[0].split('_')[1]
        meta_data = file.split('/')[-1].split('.')[0].split('_')[0]
        file_indices.append(int(file_idx))
    #finding the range of file indices to delete
    idxs_to_del_start = 1
    idxs_to_del_end = (max(file_indices)-10)
    #deleting the files
    indices_to_del = np.arange(idxs_to_del_start,idxs_to_del_end+1)
    #changing precision to have leading zeros as per file naming convention 
    indices_to_del = [format(num, "04") for num in indices_to_del]
    #iterating through each and deleting them
    for img_idx in indices_to_del:
        img_name = filepath+ "/"+str(meta_data)+"_"+str(img_idx)+".jpg"
        #deleting the file
        try:
            os.remove(img_name)
        except FileNotFoundError:
            print("{} not found".format(img_name))
    print("{} processed".format(filepath))
    
```
