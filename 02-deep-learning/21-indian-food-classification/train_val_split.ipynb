This notebook is aimed towards creating a test and train split from identical copies of dataset containing 3.5k+ images one of which is renamed as train folder and its copy as val folder inside a root directory Wonders-of-the-world

Dataset Link - https://www.kaggle.com/datasets/l33tc0d3r/indian-food-classification

```python
import os
import glob
import numpy as np
```

```python
root_dir= "/home/suraj/ClickUp/Jan-Feb/data/Food Classification/"
```

```python
# renaming images numerically for train
file_ext = "jpg"

subfolders = [ f.path for f in os.scandir(root_dir+"/train") if f.is_dir() ]
for subfolder in subfolders:
    i=0
    for filename in (glob.glob(subfolder+"/*.*")):
        #print(filename)
        i+=1
        #i = str(i)
        if len(str(i))==1:
            idx = "000"+str(i)
        elif len(str(i))==2:
            idx = "00"+str(i)
        elif len(str(i))==3:
            idx = "0"+str(i)
        else:
            idx=str(i)
        #print(filename, subfolder+"/"+str(idx) + "." + file_ext)
        os.rename(filename, subfolder+"/"+str(idx) + "." + file_ext)
        print("{} processed".format(filename))
```

```python
# renaming images numerically for val
file_ext = "jpg"

subfolders = [ f.path for f in os.scandir(root_dir+"/val") if f.is_dir() ]
for subfolder in subfolders:
    i=0
    for filename in (glob.glob(subfolder+"/*.*")):
        #print(filename)
        i+=1
        #i = str(i)
        if len(str(i))==1:
            idx = "000"+str(i)
        elif len(str(i))==2:
            idx = "00"+str(i)
        elif len(str(i))==3:
            idx = "0"+str(i)
        else:
            idx=str(i)
        #print(filename, subfolder+"/"+str(idx) + "." + file_ext)
        os.rename(filename, subfolder+"/"+str(idx) + "." + file_ext)
        print("{} processed".format(filename))
```

```python
#deleting last 10 images from each folder in train data
for filepath in glob.glob(root_dir+"/train/*"):
    file_indices=[]
    
    # finding list of last 10 images from each folder that needs to be deleted from train set 
    for file in glob.glob(filepath+"/*.jpg"):
        file_idx = file.split('/')[-1].split('.')[0]
        file_indices.append(int(file_idx))
    #finding the range of file indices to delete
    idxs_to_del_start = (max(file_indices)-10)
    idxs_to_del_end = (max(file_indices))
    #deleting the files
    indices_to_del = np.arange(idxs_to_del_start,idxs_to_del_end+1)
    #changing precision to have leading zeros as per file naming convention 
    indices_to_del = [format(num, "04") for num in indices_to_del]
    #iterating through each and deleting them
    for img_idx in indices_to_del:
        img_name = filepath+ "/"+str(img_idx)+".jpg"
        #deleting the file
        try:
            os.remove(img_name)
        except FileNotFoundError:
            print("{} not found".format(img_name))
    print("{} processed".format(filepath))
    
```

```python
#deleting last 10 images from each folder in train data
for filepath in glob.glob(root_dir+"/val/*"):
    file_indices=[]
    
    # finding list of last 10 images from each folder that needs to be deleted from train set 
    for file in glob.glob(filepath+"/*.jpg"):
        file_idx = file.split('/')[-1].split('.')[0]
        file_indices.append(int(file_idx))
    #finding the range of file indices to delete
    idxs_to_del_start = 1
    idxs_to_del_end = (max(file_indices)-10)
    #deleting the files
    indices_to_del = np.arange(idxs_to_del_start,idxs_to_del_end+1)
    #changing precision to have leading zeros as per file naming convention 
    indices_to_del = [format(num, "04") for num in indices_to_del]
    #iterating through each and deleting them
    for img_idx in indices_to_del:
        img_name = filepath+ "/"+str(img_idx)+".jpg"
        #deleting the file
        try:
            os.remove(img_name)
        except FileNotFoundError:
            print("{} not found".format(img_name))
    print("{} processed".format(filepath))
    
```
