> Optimization algorithms

Optimization algorithms are used in machine learning to find the best parameters or weights for a model and minimize the loss function.

1. Gradient Descent: Gradient descent is a widely used optimization algorithm that iteratively updates model parameters based on the negative gradient of the loss function with respect to the parameters. Here is an example code for gradient descent:

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Generate synthetic data
np.random.seed(0)
X = np.random.rand(100, 1)
y = 2 * X + 1 + 0.1 * np.random.randn(100, 1)
```

```python
# Define the cost function
def cost_function(X, y, theta):
    m = len(y)
    J = np.sum((X.dot(theta) - y) ** 2) / (2 * m)
    return J
```

```python
# Gradient Descent Algorithm
def gradient_descent(X, y, alpha=0.01, num_iterations=1000):
    m = len(y)
    theta = np.zeros((2, 1))
    J_history = np.zeros(num_iterations)

    for i in range(num_iterations):
        theta = theta - (alpha / m) * X.T.dot(X.dot(theta) - y)
        J_history[i] = cost_function(X, y, theta)

    return theta, J_history
```

```python
# Prepare the data
X_b = np.c_[np.ones((100, 1)), X]  # Add bias term
theta_init = np.random.randn(2, 1)  # Initial theta values
```

```python
# Run gradient descent
theta_optimal, J_history = gradient_descent(X_b, y, alpha=0.01, num_iterations=1000)
```

```python
# Plot the data
plt.figure(figsize=(10, 6))
plt.scatter(X, y, color='blue', label='Data Points')
plt.plot(X, X_b.dot(theta_init), color='gray', label='Initial Fit')
plt.plot(X, X_b.dot(theta_optimal), color='red', label='Optimal Fit')
plt.xlabel('X')
plt.ylabel('y')
plt.title('Gradient Descent Optimization')
plt.legend(loc='best')
plt.tight_layout()
plt.show()
```

```python
# Plot the cost function over iterations
plt.figure(figsize=(10, 6))
plt.plot(range(1, len(J_history)+1), J_history, color='blue')
plt.xlabel('Iterations')
plt.ylabel('Cost Function')
plt.title('Gradient Descent: Cost Function vs. Iterations')
plt.tight_layout()
plt.show()
```

2. Stochastic Gradient Descent (SGD): SGD is a variant of gradient descent that randomly selects a single data point to compute the gradient and update the model parameters. This is faster than gradient descent, but can make a bigger difference.

3. Mini-batch gradient descent: Mini-batch gradient descent is a variant of gradient descent that randomly selects a small batch of data points to compute the gradient and update the model parameters. This creates a balance between SGD efficiency and slope descent stability.
