#### About

> Linear Regression

Linear Regression is a commonly used statistical technique for modeling the relationship between a dependent variable and one or more independent variables. 

The goal of Linear Regression is to fit a linear model to a set of data points, represented as (x, y) pairs, where x is the input or independent variable, and y is the output or dependent variable. The linear model is represented by the equation:

y = mx + b

where m is the slope of the line, and b is the y-intercept. The slope m represents the change in the dependent variable y for each unit change in the independent variable x, and the y-intercept b represents the value of y when x is 0.

##### Cost Function

To estimate the optimal values of the model parameters m and b, we often need a way to measure the performance of the model. The Cost Function, also known as the Loss Function or Objective Function, is used to quantify the difference between the predicted values of the model and the actual values of the data points.

Commonly used cost function for linear regression is the mean squared error(MSE)

MSE = (1/n) * summation(1 to n)((y_pred_i - y_actual_i)^2)

where n is the number of data points, y_pred_i is the predicted value of y by the model, and y_actual_i is the actual value of y from the data for the ith sample.

##### Gradient Descent

To find the optimal values of m and b that minimize the cost function, we can use an optimization algorithm called Gradient Descent. Gradient Descent is an iterative algorithm that adjusts the values of the parameters in the direction of the negative gradient of the cost function, until a minimum is reached.

The update rules for m and b in Gradient Descent are given by:


m = m - lr * (1/n) * summation(1 to n)((y_pred_i - y_actual_i) * x)


b = b - lr * (1/n) * summation(1 to n)((y_pred_i - y_actual_i)

where  lr is the learning_rate









```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(0)
X = 2 * np.random.rand(100, 1)
y = 4 + 3 * X + np.random.randn(100, 1)
```

```python
# Plot the data
plt.scatter(X, y)
plt.xlabel('X')
plt.ylabel('y')
plt.title('Linear Regression Data')
plt.show()
```

```python
# Initialize model parameters
m = 0  # slope
b = 0  # y-intercept
```

```python
# Perform Gradient Descent
learning_rate = 0.01
num_iterations = 1000
n = len(X)
for i in range(num_iterations):
    y_pred = m * X + b
    mse = np.mean((y_pred - y)**2)
    dm = (1/n) * np.dot(X.T, (y_pred - y))
    db = (1/n) * np.sum(y_pred - y)
    m -= learning_rate
if i % 100 == 0:
    print(f'Iteration: {i}, MSE: {mse:.4f}')

print(f'Final Parameters: m = {m:.4f}, b = {b:.4f}')

```

```python
#plot the data and the learned regression line
y_pred = m * X + b
plt.scatter(X, y)
plt.plot(X, y_pred, color='red', label='Linear Regression')
plt.xlabel('X')
plt.ylabel('y')
plt.title('Linear Regression')
plt.legend()
plt.show()
```
