## About
Text extraction from images using PyTesseract.

Kaggle Dataset - https://www.kaggle.com/datasets/manishthem/text-extraction-for-ocr

```python
from PIL import Image
import pytesseract
import glob
import os
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
dataset_path = "/home/suraj/ClickUp/Jan-Feb/data/ImageAndXML_Data/"
for image_file in glob.glob(dataset_path+"/*.tif"):
    image = cv2.imread(image_file)
    print(image.shape)
    text = pytesseract.image_to_string(image)
    plt.imshow(image)
    plt.show()
    print(text)
    break
```
