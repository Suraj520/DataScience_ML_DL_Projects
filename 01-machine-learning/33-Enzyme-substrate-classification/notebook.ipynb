<div style="background-color:#A8B4F6; color:#19180F; font-size:30px; font-family:Arial; padding:10px; border: 5px solid #19180F; border-radius:10px"> ‚úçÔ∏è Overview</div>
<div style="background-color:#c5d5ff; color:#19180F; font-size:15px; font-family:Arial; padding:10px; border: 5px solid #19180F; border-radius:10px">
Following is a description of the properties of the enzyme substrate dataset, including the meaning and method of measurement:<br><br>
1. ID: This feature is a unique identifier for each board and has no specific meaning for the prediction task. It acts as an identifier to distinguish different substrates.<br><br>
2. BertzCT: The BertzCT feature represents a measure of the topological complexity of the board. Measures the complexity of molecular structures based on their atomic connectivity patterns. A higher value indicates a more complex structure.<br><br>
3. Chi1, Chi1n, Chi1v, Chi2n, Chi2v, Chi3v, Chi4n: These features are kappa shape indices that describe the molecular shape and connectivity pattern of the substrate. The kappa shape index is derived from the graph theory concept of molecular connectivity and provides information on the branching, cycling and overall shape of a molecule. They are computed based on the adjacency matrix of the molecular graph.<br><br>
4. EState_VSA1: EState_VSA1 is the electrical topology state VSA descriptor (VolSurf). We quantify the electrostatic properties of the substrate by considering the distribution of partial charges on the atoms. The electrical topology state index provides information about the atom's electrostatic environment and its influence on the molecular properties. <br><br>
5. EState_VSA2: No description is provided for this function, so without more information its exact meaning and how it is measured is unknown.<br><br>
6. ExactMolWt: ExactMolWt gives the exact molecular weight of the substrate. It is measured as the sum of the atomic weights of all atoms in the molecule, including hydrogen atoms. Molecular weight is a fundamental property that influences various physical and chemical properties of substrates.<br><br>
7. FpDensityMorgan1, FpDensityMorgan2, FpDensityMorgan3: These features represent the density of Morgan fingerprints, a type of molecular fingerprint used for molecular similarity calculations. Morgan fingerprints encode the presence of intramolecular substructures (often called circular substructures or Morgan circles). The density of these fingerprints reflects the abundance and diversity of substructures present on the substrate. <br><br>
8. HallKierAlpha: HallKierAlpha is the Hallkier alpha shape index that characterizes the shape of the molecule. It is based on graph theory concepts and computes the topological complexity of molecular graphs. The Alpha Shape Index measures the shape of a molecule taking into account the relative positions of atoms and their connectivity.<br><br>
9. HeavyAtomMolWt: HeavyAtomMolWt represents the molecular weight of the heavy (non-hydrogen) atoms present in the substrate and provides information about the size and complexity of the non-hydrogen components of the molecule.<br><br>
10. Kappa3: Kappa3 is the cubic kappa shape index. Measure molecular shapes and branching patterns by considering paths of specified length on the molecular diagram. A higher value of the kappa shape index indicates a more complex and branched molecular structure.<br><br>
11. MaxAbsEStateIndex: MaxAbsEStateIndex is the maximum absolute electrical topological state index. Measures the maximum absolute value of the electrostatic properties of the atoms in the substrate. This provides information about the most extreme electrostatic environment within the molecule.<br><br>
12. MinEStateIndex: MinEStateIndex is the minimum electrical topology state index. Measures the minimum electrostatic properties of the atoms in the substrate. This provides information about the most extreme electrostatic environment within the molecule.<br><br>
13. NumHeteroatoms: NumHeteroatoms indicates the number of heteroatoms (atoms other than carbon and hydrogen) in the substrate. Heteroatoms can have a significant effect on reactivity,This feature is relevant for predicting the properties of enzyme substrates for analyzing functional properties of molecules.<br><br>
14. PEOE_VSA10, PEOE_VSA14, PEOE_VSA6, PEOE_VSA7, PEOE_VSA8: These functions are the electrical topology state partial charge VSA descriptors (VolSurf). These record the distribution of partial charges on the atoms and quantify the electrostatic properties of the substrate. The VolSurf approach combines molecular surface properties and partial charges to describe the electrostatic environment around an atom.<br><br> 
15. SMR_VSA10, SMR_VSA5: These features represent the solvent accessible surface of the VSA descriptor (VolSurf) for polar atoms. They quantify the polar surface area of substrates, paying particular attention to the solvent-accessible surface area. These descriptors capture the polarity and hydrophilicity of molecules.<br><br>
16 SlogP_VSA3: SlogP_VSA3 is the SlogP VSA (VolSurf) descriptor. It represents the contribution of a particular substructure to the molecular partition coefficient (SlogP), a measure of hydrophobicity. This descriptor provides information about the hydrophobic regions or functional groups present on the substrate.<br><br>
17. VSA_ESTate9: VSA_ESTate9 is the electrical topology state VSA descriptor (VolSurf). We characterize the electrostatic properties of the substrate by considering the distribution of partial charges on the atoms. The electrical topology state index provides information about the atom's electrostatic environment and its influence on the molecular properties.<br><br>
18. fr_COO, fr_COO2: These features represent the number of carboxylic acid functional groups present on the substrate. B. Acidity and ability to form hydrogen bonds. The presence of these functional groups can affect enzyme-substrate interactions and reactivities.<br><br>
19. EC1, EC2, EC3, EC4, EC5, EC6: These features are enzyme commission numbers that classify the type of enzymatic reaction associated with the substrate. Each EC number consists of four numbers separated by periods representing a hierarchical classification of the enzyme's substrate specificity.<br><br>
The challenge in this contest is to predict EC1 and EC2 values that represent broad categories of enzymatic reactions. These features capture different aspects of the molecular structure, electrostatic properties, shape, and functional groups present in enzyme substrates. Understanding their importance and how they are measured helps to gain insight into the selection of relevant traits, building predictive models, and the relationships between molecular properties and enzyme-substrate interactions.<br><br>

```python
def plot_heatmap(df, title, figsize=(30,30)):
    # Create a mask for the diagonal elements
    mask = np.zeros_like(df.astype(float).corr())
    mask[np.triu_indices_from(mask)] = True

    # Set the colormap
    colormap = 'magma'  

    # Set the figure size
    plt.figure(figsize=figsize)

    # Set the title and font properties
    plt.title('Correlation Marix for {}'.format(title), fontweight='bold', y=1.02, size=20)

    # Plot the heatmap with the masked diagonal elements
    sns.heatmap(df.astype(float).corr(), linewidths=0.5, vmax=1.0, vmin=-1.0,
                square=True, cmap=colormap, linecolor='white', annot=True, annot_kws={"size": 13, "weight": "bold"},
                mask=mask)
```

<div style="background-color:#A8B4F6; color:#19180F; font-size:30px; font-family:Arial; padding:10px; border: 5px solid #19180F; border-radius:10px"> ‚úçÔ∏è 4. Model Selection</div>

```python
# Create a StandardScaler instance
scaler = StandardScaler()

# Scale the numerical features in X_train
numerical_features = X_train.select_dtypes(include=np.number).columns
X_train_scaled = X_train.copy()
X_train_scaled[numerical_features] = scaler.fit_transform(X_train[numerical_features])

# Scale the numerical features in X_test
X_test_scaled = X_test.copy()
X_test_scaled[numerical_features] = scaler.transform(X_test[numerical_features])
```

```python
# Check for NaN values in X_train_scaled
print(X_train_scaled.isna().sum())
```

```python
y_train
```

```python
# Create an imputer instance
imputer = SimpleImputer(strategy='mean')  # You can choose a different strategy if desired

# Fit the imputer on the training data and transform the numerical features
X_train_scaled[numerical_features] = imputer.fit_transform(X_train[numerical_features])

# Transform the numerical features of the test data
X_test_scaled[numerical_features] = imputer.transform(X_test[numerical_features])
```

```python
# Define the estimators for stacking
stacking_estimators = [
    ("Logistic Regression", LogisticRegression(max_iter=1000)),
    ("XGBoost", XGBClassifier(tree_method='gpu_hist', gpu_id=0)),
    ("LightGBM", LGBMClassifier(device='gpu')),
    ("CatBoost", CatBoostClassifier(verbose=False)),
    ("AdaBoost", AdaBoostClassifier()),
    ("Bagging", BaggingClassifier())
]

# Define the estimators for voting
voting_estimators = [
    ("Logistic Regression", LogisticRegression(max_iter=1000)),
    ("Random Forest", RandomForestClassifier()),
    ("XGBoost", XGBClassifier(tree_method='gpu_hist', gpu_id=0)),
    ("LightGBM", LGBMClassifier(device='gpu')),
    ("CatBoost", CatBoostClassifier(verbose=False)),
    ("AdaBoost", AdaBoostClassifier()),
    ("Bagging", BaggingClassifier())
]

# Update the models list
models = [
    ("Logistic Regression", LogisticRegression(max_iter=1000)),
    ("Lasso", Lasso()),
    ("Ridge", Ridge()),
    ("Elastic Net", ElasticNet()),
    ("SVM", SVC(probability=True)),
    ("Decision Tree", DecisionTreeClassifier()),
    ("Random Forest", RandomForestClassifier()),
    ("XGBoost", XGBClassifier(tree_method='gpu_hist')),
    ("LightGBM", LGBMClassifier(device='gpu', gpu_id=0)),
    ("CatBoost", CatBoostClassifier(verbose=False)),
    ("AdaBoost", AdaBoostClassifier()),
    ("Bagging", BaggingClassifier()),
    ("Stacking", StackingClassifier(estimators=stacking_estimators)),
    ("Voting", VotingClassifier(estimators=voting_estimators, voting='soft'))  # Use soft voting
]
```

```python
# Train and evaluate models using cross-validation
roc_auc_scores = []
for name, model in models:
    model_scores = []
    for target_column in y_train.columns:
        target = y_train[target_column]
        scores = cross_val_score(model, X_train_scaled, target, cv=5, scoring='roc_auc', n_jobs=-1)
        avg_score = np.mean(scores)
        model_scores.append(avg_score)
    roc_auc_scores.append(model_scores)
    print(f"{name}: ROC AUC - {model_scores}")
```

```python
# Convert roc_auc_scores to a numpy array
roc_auc_scores = np.array(roc_auc_scores)

# Plot the ROC AUC scores
plt.figure(figsize=(12, 6))
plt.bar(range(len(models)), roc_auc_scores.mean(axis=1), tick_label=[name for name, _ in models])
plt.xticks(rotation=45)
plt.ylabel("ROC AUC")
plt.title("Average ROC AUC Scores for Different Models")
plt.ylim([0, 1])
for i, v in enumerate(roc_auc_scores.mean(axis=1)):
    plt.text(i, v, f"{v:.3f}", ha="center", va="bottom")
plt.show()
```

```python
# Select the top-performing models based on ROC AUC scores
top_models_indices = roc_auc_scores.mean(axis=1).argsort()[-3:][::-1]  # Select top 3 models
top_models = [models[i] for i in top_models_indices]
print(top_models)
```


<div style="background-color:#A8B4F6; color:#19180F; font-size:30px; font-family:Arial; padding:10px; border: 5px solid #19180F; border-radius:10px"> ‚úçÔ∏è 5. Hyperparameter tuning</div>

```python
best_models = []
# Define the model to be tuned
models = [
    ("Stacking", StackingClassifier(estimators=stacking_estimators))
]

roc_auc_scores = []
for name, model in models:
    model_scores = []
    for target_column in y_train.columns:
        target = y_train[target_column]

        def objective(trial):
            # Define the model
            estimators = [
                ("Logistic Regression", LogisticRegression(max_iter=1000)),
                ("XGBoost", XGBClassifier()),
                ("LightGBM", LGBMClassifier(device='gpu')),
                ("CatBoost", CatBoostClassifier(verbose=False)),
                ("AdaBoost", AdaBoostClassifier()),
                ("Bagging", BaggingClassifier())
            ]

            model = StackingClassifier(estimators=estimators, stack_method='predict_proba')

            # Define the hyperparameters to optimize
            params = {
                "LightGBM__learning_rate": trial.suggest_float("LightGBM__learning_rate", 0.01, 0.1, log=True),
                "LightGBM__max_depth": trial.suggest_int("LightGBM__max_depth", 3, 10),
                "CatBoost__learning_rate": trial.suggest_float("CatBoost__learning_rate", 0.01, 0.1, step=0.01),
                "CatBoost__depth": trial.suggest_int("CatBoost__depth", 3, 10),
                "AdaBoost__learning_rate": trial.suggest_float("AdaBoost__learning_rate", 0.01, 0.1, log=True),
                "AdaBoost__n_estimators": trial.suggest_int("AdaBoost__n_estimators", 50, 200),
                "Bagging__n_estimators": trial.suggest_int("Bagging__n_estimators", 50, 200)
            }

            # Set the hyperparameters for the model
            model.set_params(**params)

            # Evaluate the model using cross-validation
            scores = cross_val_score(model, X_train_scaled, target, cv=5, scoring='roc_auc', n_jobs=-1)

            # Return the mean score as the objective value
            return scores.mean()

        # Create an Optuna study and optimize the objective function
        study = optuna.create_study(direction="maximize")
        study.optimize(objective, n_trials=2)

        # Get the best hyperparameters and model
        best_params = study.best_params
        best_model = StackingClassifier(estimators=stacking_estimators, stack_method='predict_proba')
        best_model.set_params(**best_params)
        
        best_models.append((name, best_model, best_params))

        scores = cross_val_score(best_model, X_train_scaled, target, cv=5, scoring='roc_auc', n_jobs=-1)
        avg_score = np.mean(scores)
        model_scores.append(avg_score)

    roc_auc_scores.append(model_scores)
    print(f"{name}: ROC AUC - {model_scores}")
```
