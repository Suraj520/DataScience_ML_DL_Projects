#### About

> Sampling distributions

In statistics, a sampling distribution is the probability distribution of a statistic based on a random sample. In other words, it is the distribution of all possible values of a statistic that could be obtained from samples of a given size taken from a population.

Sampling distributions are important in statistical inference, as they allow us to make inferences about a population based on a sample. By understanding the properties of sampling distributions, we can estimate population parameters with a certain degree of confidence and test hypotheses about the population.




- Example :

Suppose we want to estimate the mean weight of all apples produced by a farm. We take a random sample of 100 apples from the farm and measure their weights in grams. We repeat this process many times and calculate the mean weight of each sample. The distribution of these sample means is the sampling distribution of the mean.



```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Generate population of apple weights (normal distribution with mean 150g and standard deviation 20g)
population = np.random.normal(150, 20, 100000)
```

```python
sample_means = []
for i in range(1000):
    sample = np.random.choice(population, 100)
    sample_means.append(np.mean(sample))

# Take random samples of size 100 and calculate their means
```

```python
# Plot histogram of sample means
plt.hist(sample_means, bins=30)
plt.xlabel('Mean weight (g)')
plt.ylabel('Frequency')
plt.title('Sampling Distribution of the Mean')
plt.show()
```

The resulting plot shows the sampling distribution of the mean, which is approximately normal due to the central limit theorem. The mean of the sampling distribution is equal to the population mean, and the standard deviation is equal to the population standard deviation divided by the square root of the sample size.
