> Cumulative distribution function

The Cumulative Distribution Function (CDF) of a random variable is a function that gives the probability that the variable takes a value less than or equal to a given value. In other words, it is the probability distribution of the random variable X.

Use cases

1. Probability calculations: The CDF can be used to calculate the probability of an event occurring between two values.

2. Hypothesis testing: The CDF can be used to calculate critical values for hypothesis testing.

3. Data visualization: The CDF can be used to visualize the distribution of a random variable.

Suppose we have a dataset of the weights of a group of people, and we want to calculate the CDF of the weights:



```python
import numpy as np
import matplotlib.pyplot as plt

# Generate some rweight data
weight_data = np.random.normal(loc=70, scale=10, size=1000)
```

```python
# Sort the data in ascending order
data_sorted = np.sort(weight_data)
```

```python

# Calculate the CDF
cdf = np.cumsum(data_sorted) / np.sum(data_sorted)
```

```python
# Plot the CDF
plt.plot(data_sorted, cdf)
plt.xlabel('Weight')
plt.ylabel('Cumulative Probability')
plt.show()






```
