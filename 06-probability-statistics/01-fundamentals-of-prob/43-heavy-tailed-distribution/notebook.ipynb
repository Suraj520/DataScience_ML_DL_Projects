> Heavy Tailed Distribution

In probability theory and statistics, a heavy-tailed distribution refers to a probability distribution whose tail is larger than that of the normal distribution. Such distributions have a high probability of generating extreme values or outliers that lie far from the mean.



```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import powerlaw
```

```python
alpha = 2.5  # shape parameter
xmin = 1.0   # minimum value
n = 1000     # sample size
s = powerlaw.rvs(alpha, size=n) + xmin
#gen data from power law
```

```python
# Plot the distribution
fig, ax = plt.subplots()
ax.hist(s, bins=50, density=True, log=True, alpha=0.5)
ax.set(xlabel='Value', ylabel='Frequency',
       title='Power Law Distribution (alpha={:.1f}, xmin={:.1f})'.format(alpha, xmin))
plt.show()
```
